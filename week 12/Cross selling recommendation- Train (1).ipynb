## Visualization of the statistical analysis

```python
df
```

```python
df[df['ind_nuevo']==1]['ind_nuevo'].count()/df[df.columns[0]].count()*100
```

```python
fig, ax = plt.subplots()
new = df[df['ind_nuevo']==1]['ind_nuevo'].count()/df[df.columns[0]].count()*100
old = (1-df[df['ind_nuevo']==1]['ind_nuevo'].count()/df[df.columns[0]].count())*100
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = '#909090'
plt.rcParams['axes.labelcolor']= '#909090'
plt.rcParams['xtick.color'] = '#909090'
plt.rcParams['ytick.color'] = '#909090'
plt.rcParams['font.size']=12
labels = ['New Customers', 
         'Old Customers']
color_palette_list = ['#ADD8E6','#009ACD', '#63D1F4', '#0EBFE9',   
                      '#C1F0F6', '#0099CC']
percentages = [new,old]
explode=(0.1,0)
ax.pie(percentages, explode=explode, labels=labels,  
       colors=color_palette_list[0:2], autopct='%1.0f%%', 
       shadow=False, startangle=0,   
       pctdistance=1.2,labeldistance=1.4)
ax.axis('equal')
ax.set_title("Percentage of new users")
ax.legend(frameon=False, bbox_to_anchor=(1.5,0.8))
```

Based on the graph above, it is clear that the proportion of new customers is relatively tiny by comparison to the proportion of old customers. Therefore, one of the key focus for the company is to expand the new customers in order to expand the cross selling business.

```python
# Old Customers
df[df['ind_nuevo']==0]['age'].describe()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.boxplot(df[df['ind_nuevo']==0]['age'])
```

Based on the two boxplot and describe above, the group of old customers has bigger age span than the group of new customers has. Other than that, two groups of customers have the different medians and the IQRs, so the majorities of customers' age from  20 to 40 is the main force of the new customers. Hence, the company can put more attention on the product for 20 to 40 age for attracting new customers in the business of cross selling. 

```python
new_first_quantile = df[df['ind_nuevo']==1]['age'].quantile(0.25)
new_first_quantile
```

```python
fig, ax = plt.subplots()
new_first_quantile = df[df['ind_nuevo']==1]['age'].quantile(0.25)
new_third_quantile = df[df['ind_nuevo']==1]['age'].quantile(0.75)
new_active = df[(df['ind_nuevo']==1) & (df['age']>=new_first_quantile) & (df['age']<=new_third_quantile) & (df['ind_actividad_cliente']==1)]['ind_actividad_cliente'].count()
new_total = df[(df['ind_nuevo']==1) & (df['age']>=new_first_quantile) & (df['age']<=new_third_quantile)]['ind_actividad_cliente'].count()
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = '#909090'
plt.rcParams['axes.labelcolor']= '#909090'
plt.rcParams['xtick.color'] = '#909090'
plt.rcParams['ytick.color'] = '#909090'
plt.rcParams['font.size']=12
labels = ['New Active Customers', 
         'New Inactive Customers']
color_palette_list = ['#009ACD','#ADD8E6', '#63D1F4', '#0EBFE9',   
                      '#C1F0F6', '#0099CC']
percentages = [new_active,new_total - new_active]
explode=(0.1,0)
ax.pie(percentages, explode=explode, labels=labels,  
       colors=color_palette_list[0:2], autopct='%1.0f%%', 
       shadow=False, startangle=0,   
       pctdistance=1.2,labeldistance=1.4)
ax.axis('equal')
ax.set_title("Percentage of new active users\n", pad = 20, loc = 'center')
ax.legend(frameon=False, bbox_to_anchor=(1.5,0.8))
```

```python
fig, ax = plt.subplots()
old_first_quantile = df[df['ind_nuevo']==0]['age'].quantile(0.25)
old_third_quantile = df[df['ind_nuevo']==0]['age'].quantile(0.75)
old_active = df[(df['ind_nuevo']==0) & (df['age']>=old_first_quantile) & (df['age']<=old_third_quantile) & (df['ind_actividad_cliente']==1)]['ind_actividad_cliente'].count()
old_total = df[(df['ind_nuevo']==0) & (df['age']>=old_first_quantile) & (df['age']<=old_third_quantile)]['ind_actividad_cliente'].count()
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = '#909090'
plt.rcParams['axes.labelcolor']= '#909090'
plt.rcParams['xtick.color'] = '#909090'
plt.rcParams['ytick.color'] = '#909090'
plt.rcParams['font.size']=12
labels = ['Old Active Customers', 
         'Old Inactive Customers']
color_palette_list = ['#009ACD','#ADD8E6', '#63D1F4', '#0EBFE9',   
                      '#C1F0F6', '#0099CC']
percentages = [old_active,old_total - old_active]
explode=(0.1,0)
ax.pie(percentages, explode=explode, labels=labels,  
       colors=color_palette_list[0:2], autopct='%1.0f%%', 
       shadow=False, startangle=0,   
       pctdistance=1.2,labeldistance=1.4)
ax.axis('equal')
ax.set_title("Percentage of new active users\n\n", pad = 20, loc = 'right')
ax.legend(frameon=False, bbox_to_anchor=(1.5,0.8))
```

Based on two pie charts above, which display the ratio of active and inactve customers in the groups of new and old customers main aged group, it obvious that the amount of active customers is dominanted that of inactive customers in the group of new customers. However, in the group of old customers, the amount of active customers is equal to that of inactive customers. Therefore, expanding the business for new customers is the key to success in cross selling.
